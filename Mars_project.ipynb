## Importing of Packages

```python
import pandas
import numpy as np
import matplotlib.pyplot as plt

from pandas import DataFrame
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv("cleaned-mars-weather.csv")
```

```python
data.min()
```

# Max_temp

```python
x = DataFrame(data, columns=['sol'])
y = DataFrame(data, columns=['max_temp'])
```

```python
regression = LinearRegression()
regression.fit(x,y)
```

```python
plt.figure(figsize=[10,10])
plt.title("Maximum_temp")
plt.xlabel("Sol")
plt.ylabel("Max_temp")

plt.xlim(10,2000)
plt.ylim(-35,15)

plt.scatter(x, y, color='red')
plt.plot(x,regression.predict(x), color='blue')

plt.show()
```

```python
regression.coef_
```

```python
regression.intercept_
```

# Max_temp function

$$ f(x) = -7.43832512 + (x * -0.00499387) $$ 

```python
def max_temp(x):
    return(-7.43832512 + (x * -0.00499387))
```

```python
#Example problems
max_temp(2005)
max_temp(3000)
```
